#### To make analysis of data in table easier, we can reshape the data into a more computer-friendly form using Pandas in Python. Pandas.melt() is one of the function to do so..
Pandas.melt() unpivots a DataFrame from wide format to long format.
melt() function is useful to message a DataFrame into a format where one or more columns are identifier variables, while all other columns, considered measured variables, are unpivoted to the row axis, leaving just two non-identifier columns, variable and value.
Syntax :



```python
import pandas as pd
```

```python
df= pd.read_csv("/Users/aryansaini/Documents/Virtual Environment/Python/Pandas/Part 3/weather.csv")
df
```

```python
#Using melt to transform the data frame

df1 = pd.melt(df, id_vars=['day'])
df1[df1['variable']=='chennai']
```

```python
df1 = pd.melt(df, id_vars=['day'], var_name='city', value_name='temperature')
df1
```

## Contingency table / Cross Tabs
### Crosstab (also known as contingency table or cross tabulation)
### is a table showing frequency distribution of one variable in 
### rows and another on columns. pandas crosstab method can be used 
### to generate these contingency tables that are extremely useful in survey and business analytics.

```python
df = pd.read_excel("/Users/aryansaini/Documents/Virtual Environment/Python/Pandas/Part 3/survey.xls")
df
```

```python
pd.crosstab(df.Sex, df.Age)
```

```python
#We can call the margin arguement to view the total 
pd.crosstab(df.Nationality, df.Handedness, margins=True)
```

```python
#If we want to have multiple columns 
pd.crosstab([df.Nationality,df.Sex, df.Age], df.Handedness ,normalize=True)
```

```python
pd.crosstab(df.Nationality, df.Handedness, normalize='index')
```

```python
pd.crosstab(df.Sex, df.Handedness)
```

```python
#Suppose now we want to find out the average age of male or females based on the contigency table
import numpy as np
pd.crosstab(df.Sex, df.Handedness, values=df.Age, aggfunc= np.average)
```
